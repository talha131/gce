# Output

## Output Directory

````python
def ensure_output_directory(output_dir: str) -> None:
    """Ensure that the output directory exists, create it if it doesn't."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        logging.info(f"Created output directory: {output_dir}")
````

## Generate CSV Files

````python
def generate_csv_files(config: Dict[str, Any]) -> None:
    # Create a timestamped folder name
    timestamp = datetime.now().strftime("%Y-%m-%d %H%M")
    output_dir = f"/content/drive/My Drive/csv_output_{timestamp}"
    ensure_output_directory(output_dir)

    for i in range(config["num_files"]):
        num_students = random.randint(config["min_students"], config["max_students"])
        num_subjects = random.randint(config["min_subjects"], config["max_subjects"])
        subjects = random.sample(subjects_pool, num_subjects)

        selected_students = random.sample(
            student_names * ((num_students // len(student_names)) + 1), num_students
        )

        marks = generate_marks(num_students, num_subjects, config)

        file_name = f"{config['file_prefix']}{config['start_file_suffix'] + i}.csv"
        csv_file_path = os.path.join(output_dir, file_name)

        try:
            with open(csv_file_path, "w", newline="") as csvfile:
                csv_writer = csv.writer(csvfile)

                header = ["Name"] + subjects
                csv_writer.writerow(header)

                for idx, student_name in enumerate(selected_students):
                    csv_writer.writerow([student_name] + marks[idx])

            logging.info(f"Successfully generated {file_name}")
        except IOError as e:
            logging.error(f"Error writing to file {file_name}: {e}")
````

# Generate the Data

## Generate Probabilities

````python
def generate_probabilities(mean_prob: float, num: int, spread_factor: float) -> np.ndarray:
    """
    Generate an array of probabilities based on a mean and spread factor.

    :param mean_prob: The mean probability
    :param num: Number of probabilities to generate
    :param spread_factor: Controls the spread of probabilities around the mean
                          Higher values increase variability, lower values decrease it
    :return: Array of probabilities
    """
    std_dev = mean_prob * spread_factor
    probs = np.clip(np.random.normal(mean_prob, std_dev, num), 0, 1)
    return probs
````

## Generate Subject Parameters

````python
def generate_subject_params(num_subjects, config):
    difficulty_weights = list(config["subject_difficulty"].values())
    difficulty_types = list(config["subject_difficulty"].keys())

    subject_difficulties = random.choices(difficulty_types, weights=difficulty_weights, k=num_subjects)

    subject_params = []
    for difficulty in subject_difficulties:
        alpha_range = config["subject_params"][difficulty]["alpha_range"]
        beta_range = config["subject_params"][difficulty]["beta_range"]

        alpha = random.uniform(*alpha_range)
        beta = random.uniform(*beta_range)

        subject_params.append((alpha, beta))

    return subject_params
````

## Generate Marks

````python
def generate_marks(num_students: int, num_subjects: int, config: Dict[str, Any]) -> List[List[str]]:
    marks_data = []

    base_absent_prob = config["absent_probability"]

    student_types = config["student_types"]
    performance_params = config["performance_params"]

    subject_params = generate_subject_params(num_subjects, config)

    for _ in range(num_students):
        student_type = np.random.choice(list(student_types.keys()), p=list(student_types.values()))
        row = []

        # Adjust absent probability based on student type
        if student_type == 'high_performer':
            absent_factor = 0.2  # High performers are 5 times less likely to be absent
        elif student_type == 'low_performer':
            absent_factor = 2.0  # Low performers are twice as likely to be absent
        else:  # moderate performer
            absent_factor = 1.0  # No change for moderate performers

        for subject in range(num_subjects):
            subject_absent_prob = min(base_absent_prob * absent_factor, 1.0)

            if random.random() < subject_absent_prob:
                row.append(config["absent"])
            else:
                subject_alpha, subject_beta = subject_params[subject]

                if student_type == 'high_performer':
                    alpha = performance_params["high"]["alpha"] + subject_alpha
                    beta = performance_params["high"]["beta"] + subject_beta / 2
                elif student_type == 'low_performer':
                    alpha = performance_params["low"]["alpha"] + subject_alpha / 2
                    beta = performance_params["low"]["beta"] + subject_beta
                else:  # moderate performer
                    alpha = performance_params["moderate"]["alpha"] + subject_alpha
                    beta = performance_params["moderate"]["beta"] + subject_beta

                # Add some randomness to create more variance
                alpha *= random.uniform(0.9, 1.1)
                beta *= random.uniform(0.9, 1.1)

                mark = int(np.random.beta(alpha, beta) * config["max_marks"])
                row.append(str(mark))

        marks_data.append(row)

    return marks_data
````

# Configuration

### `absent_probability` (0.25)

* This is the base probability for a student being absent in any given subject.
* Increasing this value (e.g., to 0.4) will result in more absences across all student types.
* Decreasing it (e.g., to 0.1) will result in fewer absences overall.

### `spread_factor` (0.75)

* This controls the variability of absence probabilities across different subjects.
* A higher value (e.g., 1.0) will create more diverse absence rates between subjects.
* A lower value (e.g., 0.5) will make absence rates more uniform across subjects.

### `student_types`

* These proportions determine the distribution of student types in the generated data.
* Adjusting these will directly affect the overall performance distribution.
* For example, increasing 'high_performer' will result in more high-scoring students overall.

### `performance_params`

* These beta distribution parameters control the shape of the score distribution for each student type.

#### For "high" performers
* Increasing alpha (e.g., to 10) will push scores even higher.
* Decreasing beta (e.g., to 1) will make very high scores more common.

#### For "low" performers

* Increasing beta (e.g., to 10) will push scores even lower.
* Decreasing alpha (e.g., to 1) will make very low scores more common.

#### For "moderate" performers

* Keeping alpha and beta equal maintains a centered distribution.
* Increasing both (e.g., to 7,7) will cluster scores more tightly around the middle.
* Decreasing both (e.g., to 3,3) will spread scores out more evenly.

## Beta Distribution Shapes:

1. alpha > beta (skewed to higher scores)
   e.g., alpha=8, beta=2

   This creates a distribution skewed towards higher scores. Most scores will be in the upper range.

```
     |    *
     |   * *
     |  *   *
     | *     *
     |*       *
   __|*         *___
     0    50    100   (Score)
```

2. alpha < beta (skewed to lower scores)
   e.g., alpha=2, beta=8

   e.g., alpha=5, beta=5

```

     |    *
     |   * *
     |  *   *
     | *     *
     |*       *
   __|*_______*___
     0    50    100   (Score)

```

4. Increasing both alpha and beta (more peaked)
   e.g., alpha=10, beta=10

   This makes the distribution more peaked around the center, meaning scores will be more clustered around the mean.

  ```
     |     *
     |    * *
     |   *   *
     |  *     *
     | *       *
   __|*_________*___
     0    50    100   (Score)

  ```

5. Decreasing both alpha and beta (more spread out)
   e.g., alpha=2, beta=2

   This spreads out the distribution, making extreme scores (both high and low) more likely.

## Validate Configuration

````python
import sys
from typing import Dict, Any

def validate_config(config: Dict[str, Any], subject_pool: list, student_names: list) -> None:
    """
    Validate the configuration dictionary.
    Raises ValueError with a descriptive message if validation fails.
    """
    # Check if max_subjects doesn't exceed the length of subject_pool
    if config['max_subjects'] > len(subject_pool):
        raise ValueError(f"max_subjects ({config['max_subjects']}) cannot exceed the number of available subjects ({len(subject_pool)})")

    # Check if max_students doesn't exceed the length of student_names
    if config['max_students'] > len(student_names):
        raise ValueError(f"max_students ({config['max_students']}) cannot exceed the number of available student names ({len(student_names)})")

    # Check if student_types probabilities sum to 1.0 (allowing for small float imprecision)
    student_type_sum = sum(config['student_types'].values())
    if not 0.99 <= student_type_sum <= 1.01:
        raise ValueError(f"Probabilities in student_types must sum to 1.0, but they sum to {student_type_sum}")

    # Check if min_students is less than or equal to max_students
    if config['min_students'] > config['max_students']:
        raise ValueError(f"min_students ({config['min_students']}) cannot be greater than max_students ({config['max_students']})")

    # Check if min_subjects is less than or equal to max_subjects
    if config['min_subjects'] > config['max_subjects']:
        raise ValueError(f"min_subjects ({config['min_subjects']}) cannot be greater than max_subjects ({config['max_subjects']})")

    # Check if absent_probability is between 0 and 1
    if not 0 <= config['absent_probability'] <= 1:
        raise ValueError(f"absent_probability ({config['absent_probability']}) must be between 0 and 1")

    # Check if spread_factor is positive
    if config['spread_factor'] <= 0:
        raise ValueError(f"spread_factor ({config['spread_factor']}) must be positive")

    # Check if performance_params have valid alpha and beta values (must be positive)
    for performance_type, params in config['performance_params'].items():
        if params['alpha'] <= 0 or params['beta'] <= 0:
            raise ValueError(f"Alpha and beta for {performance_type} performers must be positive. Current values: alpha={params['alpha']}, beta={params['beta']}")

    # Check if subject_difficulty probabilities sum to 1.0 (allowing for small float imprecision)
    difficulty_sum = sum(config['subject_difficulty'].values())
    if not 0.99 <= difficulty_sum <= 1.01:
        raise ValueError(f"Probabilities in subject_difficulty must sum to 1.0, but they sum to {difficulty_sum}")

    # Check if all necessary keys are in subject_params
    for difficulty in ['easy', 'medium', 'hard']:
        if difficulty not in config['subject_params']:
            raise ValueError(f"Missing '{difficulty}' in subject_params")
        if 'alpha_range' not in config['subject_params'][difficulty] or 'beta_range' not in config['subject_params'][difficulty]:
            raise ValueError(f"Missing 'alpha_range' or 'beta_range' for '{difficulty}' in subject_params")

    # Check if all ranges in subject_params are valid
    for difficulty, params in config['subject_params'].items():
        if params['alpha_range'][0] >= params['alpha_range'][1] or params['beta_range'][0] >= params['beta_range'][1]:
            raise ValueError(f"Invalid range for '{difficulty}' in subject_params")

    # All checks passed
    logging.info("Configuration validated successfully.")
````

````python
def main() -> None:
    try:
        # Validate the configuration
        validate_config(config, subjects_pool, student_names)

        # If validation passes, generate CSV files
        generate_csv_files(config)
    except ValueError as ve:
        # Catch and log validation errors
        logging.error(f"Configuration error: {str(ve)}")
        print(f"Error in configuration: {str(ve)}")
        sys.exit(1)
    except Exception as e:
        # Catch and log other unexpected errors
        logging.exception("An unexpected error occurred:")
        print(f"An unexpected error occurred: {str(e)}")
        sys.exit(1)

if __name__ == "__main__":
    main()
````
